```python
import pandas as pd
```

```python
# Read the csv file in pandas
cities_table = pd.read_csv("Cities.csv")
```

```python
#setting up city are the index

city_table = cities_table.set_index('City')


city_table.head()
```

```python
# Save to file
html = city_table.to_html(index=False).replace("\n","") 
html
```

```python
with open("data_table.html","w", encoding="UTF-8") as text_file:
    text_file.write(html)
```
